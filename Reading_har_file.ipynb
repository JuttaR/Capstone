```python
import json
from haralyzer import HarPage

with open('mercedes.har', 'r') as f:
    har_page = HarPage('unknown', har_data=json.loads(f.read()))
    
#filter entries to relevant types
entries = har_page.filter_entries(content_type='application/json', status_code='2.*', request_type='GET')    
print(f'Filtered Entries: {entries}\n')

#trying to extract response text from 1st entry
print(f'Response Text: {entries[0].response.text}')
```
